# Baseline Correction Approaches

## Approach 1: Asymmetric Least Squares Smoothing

```python
# Baseline correction function -source: "Asymmetric Least Squares Smoothing" 
# by P. Eilers and H. Boelens in 2005.

from scipy import sparse
from scipy.sparse.linalg import spsolve

def baseline_als(y, lam, p, niter=10):
  L = len(y)
  D = sparse.diags([1,-2,1],[0,-1,-2], shape=(L,L-2))
  w = np.ones(L)
  for i in range(niter):
    W = sparse.spdiags(w, 0, L, L)
    Z = W + lam * D.dot(D.transpose())
    z = spsolve(Z, w*y)
    w = p * (y > z) + (1-p) * (y < z)
  return z
```

```python
#Correct baseline of Ovarian Cancer mean intensities using ALSS method where lam is lamda
# and p is asymmetry parameter
baseline_corrected_oc = baseline_als(ovarian_mean, lam = 10^7, p = 0.001)

baseline_corrected_con = baseline_als(control_mean, lam = 10^7, p = 0.001)
```

```python
#Find peaks after using ALSS baseline correction with interpolation
from scipy.signal import find_peaks

plt.figure(figsize=(20,12))


#OVARIAN CANCER 

x_corr_oc = mz_ovarian
y_corr_oc = baseline_corrected_oc

f_oc_corr = scipy.interpolate.splrep(x_corr_oc, y_corr_oc, k = 3, s = 10) # s is for smoothing, k is for order of spline

signal_oc = np.array(splev(x_corr_oc,f_oc_corr))
peaks_oc, _ = find_peaks(signal_oc, height=0)

print('ovarian cancer after ALSS baseline correction with interpolation/smoothing:', len(peaks_oc))

plt.plot(signal_oc, color = "navy", label = "ovarian cancer post-ALSS baseline correction & interpolation/smoothing")
plt.plot(peaks_oc, signal_oc[peaks_oc], "x", color = "blue", label = "ovarian cancer peaks post-ALSS baseline correction & interpolation/smoothing")


#CONTROL

x_corr_con = mz_control
y_corr_con = baseline_corrected_con

f_con_corr = scipy.interpolate.splrep(x_corr_con, y_corr_con, k = 3, s = 10) # s is for smoothing, k is for order of spline

signal_con = np.array(splev(x_corr_con,f_con_corr))
peaks_con, _ = find_peaks(signal_con, height=0)

print('control after ALSS baseline correction with interpolation/smoothing:', len(peaks_con))

plt.plot(signal_con, color = "grey", label = "control post-ALSS baseline correction & interpolation/smoothing")
plt.plot(peaks_con, signal_con[peaks_con], "x", color = "black", label = "control peaks post-ALSS baseline correction & interpolation/smoothing")



plt.plot(np.zeros_like(x), "--", color= "gray")
plt.xlabel("m/z")
plt.ylabel("Intensity")
plt.legend(loc=0)
plt.show()
```

## Approach 2: Baseline subraction with polynomial regression approximation

```python
#Identify baseline using iterative polynomial regression as approximation

import peakutils
from peakutils.plot import plot as pplot

baseline_oc = peakutils.baseline(baseline_corrected_oc, deg = 8)
baseline_con = peakutils.baseline(baseline_corrected_con, deg = 8)

plt.figure(figsize=(20,12))
plt.plot(baseline_oc, color = "navy", linestyle = 'dashed', label = "baseline approximation for ovarian cancer")
plt.plot(baseline_corrected_oc, color = "navy", label = "ovarian cancer with ALSS/smoothing/interpolation")
plt.plot(baseline_con, color = "grey", linestyle = 'dashed', label = "baseline approximation for control")
plt.plot(baseline_corrected_con, color = "grey", label = "control with ALSS/smoothing/interpolation")
plt.xlabel("m/z")
plt.ylabel("Intensity")
plt.legend(loc = 0)
plt.show()
```

```python
#Subtract out baseline approximation
ovarian_cancer = baseline_corrected_oc - baseline_oc
control = baseline_corrected_con - baseline_con

#Plot resulting spectrogram
plt.figure(figsize=(20,12))
plt.plot(ovarian_cancer, color = "navy", label = "ovarian cancer")
plt.plot(control, color = "grey", label = "control")

plt.xlabel("m/z")
plt.ylabel("Intensity")
plt.legend(loc = 0)
```
