# Analysis of Educational Program Benefitting Inmates at Different Indian Prisons

Dataset Source: National Crime Records Bureau (NCRB), Govt of India has shared this dataset under Govt. Open Data License - India. NCRB has also shared prison data on their website.

The data refers to State/UT-wise details of prison inmates benefited by elementary education, adult education, high education and computer courses.

```python
#Import packages needed for analysis
import pandas as pd
import numpy as np
```

```python
#Read in data file from csv
prisoners = pd.read_csv("prisoners.csv")
```

In 2013, all of the programs did not benefit prisoners in certain prisons. This may be due to a lack of offering of the program or participation by prisons in these programs for prisoners. 

```python
prisoners_totals = prisoners.drop(columns = ['YEAR'])
prisoners_totals.head()
```

```python
#Create new column and new row for total benefitted inmates and totals for each method of education
prisoners_totals.loc[:,'total_benefitted'] = prisoners_totals.sum(axis = 1)
prisoners_totals.loc['totals',:] = prisoners_totals.sum(axis = 0)
prisoners_totals
```

### Let's investigate which factors benefit prisoners the most

First, which state has the maximum number of beneficiaries?

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
plt.figure(figsize=(20,10))

plt.bar(x = 'STATE/UT', height = 'total_benefitted', data= prisoners_totals.drop(['totals']))

plt.xlabel('State/UT')
plt.ylabel('Total Benefitted Inmates')
plt.xticks(rotation=90, size = 16)
plt.show()
```

**Answer**

Andhra Pradeesh and Uttar Pradeesh have the first and second highest respective number of benefitted inmates.

```python
labels = 'Elementary Education', 'Adult Education', 'Higher Education', 'Computer Course'
sizes = list(prisoners_totals.drop(columns = ['STATE/UT', 'total_benefitted']).iloc[-1])
plt.pie(sizes , labels = labels, autopct='%1.1f%%')
plt.show();
```

Adult education had the greatest benefit to prisoners, followed by Elementary Education, Higher Education, and completion of a computer class.
